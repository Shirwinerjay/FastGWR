```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gwr = np.array([18.0,31.07,57.89,107.72,210.04,673.0])
gwr = gwr[-1]/gwr

mgwr = np.array([888.39,1165.61,583,882,1479,6334])
mgwr = mgwr[-1]/mgwr
mgwr_2 = np.array([163.48,317.79,621.77,2429.0, 111,111])
nproc = [80,40,20,10,5,1]
```

```python
flg = plt.figure(figsize=(6, 6))

plt.plot(nproc,gwr,marker="o",label="GWR",markersize=5)
plt.plot(nproc,mgwr,marker="o",label="MGWR",markersize=5)
plt.plot(nproc,mgwr_2,marker="o",label="MGWR (estimation only)",markersize=5)
#plt.ylim(0,40)
plt.ylabel('Speed-up',fontsize=13)
plt.xlabel('Number of Processors',fontsize=13)
plt.legend(loc='upper left',fontsize=13)
#plt.title('(b)',pad=-390)

plt.tight_layout()
plt.savefig('scalability.png', format='png', dpi=900)
```
